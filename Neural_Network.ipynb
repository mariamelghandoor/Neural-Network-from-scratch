activation functions

```python
def relu(x):
    return np.maximum(0, x)

def softmax(x):
    exp_x = np.exp(x)
    return exp_x / np.sum(exp_x, axis=1, keepdims=True)

def sigmoid(x):
    try:
        return 1 / (1 + np.exp(-x))
    except RuntimeWarning:
        return 0

def tanh(x):
    return 1 - np.tanh(x)**2

def leaky_relu(x, alpha = 0.01):
   return np.maximum(alpha * x, x)

def swish(x):
    return x * sigmoid(x)
```

```python
############3same but clip for nan problem
import numpy as np

def clip_gradients(grad, threshold):
    clipped_grad = np.clip(grad, -threshold, threshold)
    return clipped_grad

def NN_training_function_softmax_clip(x, y, features, hidden_layer, output_layer, iterations, stop_number, step, gradient_clip_threshold=1.0):
    # set random data to begin with
    bias_input = 1
    bias_input_weight = 0.1
    bias_hidden = 2
    bias_hidden_weight = 0.3

    #set weights random
    weights_between_input_and_hidden = np.random.rand(features, hidden_layer) * 0.1
    weights_between_hidden_and_output = np.random.rand(hidden_layer, output_layer) * 0.2

    #stuff i will return
    best_weights_input_hidden = None
    best_weights_hidden_output = None
    best_bias_input_weight = bias_input_weight
    best_bias_hidden_weight = bias_hidden_weight
    best_mse = np.inf

    for i in range(iterations):
        # forward
        hidden_layer_outputs = softmax(np.dot(x, weights_between_input_and_hidden) + bias_input_weight)
        output_layer_outputs = softmax(np.dot(hidden_layer_outputs, weights_between_hidden_and_output) + bias_hidden_weight)

        # backward
        alpha_output = y - output_layer_outputs
        alpha_weights_output = np.dot(hidden_layer_outputs.T, alpha_output)
        alpha_hidden = np.dot(alpha_output, weights_between_hidden_and_output.T)
        alpha_weights_hidden = np.dot(x.T, alpha_hidden)

        # the clipping proccess to prevent nan try 20
        alpha_weights_output = clip_gradients(alpha_weights_output, gradient_clip_threshold)
        alpha_weights_hidden = clip_gradients(alpha_weights_hidden, gradient_clip_threshold)

        # update
        weights_between_hidden_and_output += step * alpha_weights_output
        bias_hidden_weight += step * bias_hidden_weight
        weights_between_input_and_hidden += step * alpha_weights_hidden
        bias_input_weight += step * bias_input_weight

        # mse calculations
        error = y - output_layer_outputs
        mse = np.mean(np.square(error))

        # update best weights and biases to return baest results
        if mse < best_mse:
            best_mse = mse
            best_weights_input_hidden = np.copy(weights_between_input_and_hidden)
            best_weights_hidden_output = np.copy(weights_between_hidden_and_output)
            best_bias_input_weight = bias_input_weight
            best_bias_hidden_weight = bias_hidden_weight


        print(f"The mean squared error in iteration {i} is: {mse * 100}")


        diff = np.abs(error)
        if np.all(diff < stop_number):
            print("Successfully trained :)")
            return best_bias_hidden_weight, best_bias_input_weight, best_weights_input_hidden , best_weights_hidden_output


        if np.isnan(output_layer_outputs).any():
            break

    print(f"With {iterations} iterations, the output reached:")
    print(output_layer_outputs)
    print("Training accuracy:", 100 - mse * 100, "%")
    print("Best MSE:", best_mse*100)
    return best_bias_hidden_weight, best_bias_input_weight, best_weights_input_hidden , best_weights_hidden_output
```
